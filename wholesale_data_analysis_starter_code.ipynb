## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
# Calculate the subtotal for each line
client_df['line_subtotal'] = client_df['unit_price'] * client_df['qty']

# Display the first few rows to verify the new column
print(client_df[['line_subtotal']].head())
```

```python
client_df.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.

client_df['shipping_price'] = client_df.apply(
    lambda row: 7 * row['unit_weight'] * row['qty'] if row['unit_weight'] > 50 else 10 * row['unit_weight'] * row['qty'], 
    axis=1
)

print(client_df[['shipping_price']].head())
```

```python
client_df.head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%

client_df['total_price'] = (client_df['line_subtotal'] + client_df['shipping_price']) * (1 + 0.0925)


# Display the first few rows to verify the new column
print(client_df[['line_subtotal', 'shipping_price', 'total_price']].head())

13,188
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).

# Create a new column for the cost of each line
client_df['line_cost'] = (client_df['unit_cost'] * client_df['qty']) + client_df['shipping_price']

# Display the first few rows to verify the new column
print(client_df[['unit_cost', 'qty', 'shipping_price', 'line_cost']].head())
```

```python
# Create a column for the profit of each line using line cost and line price

# Create a new column for the profit of each line
client_df['line_profit'] = client_df['total_price'] - client_df['line_cost']

# Display the first few rows to verify the new column
print(client_df[['total_price', 'line_cost', 'line_profit']].head())

```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.

# Top 5 clients' IDs from Part 1
top_5_clients_ids = [33615, 66037, 46820, 38378, 24741]

# Filter the data for these clients and sum their total spending
top_clients_spending = client_df[client_df['client_id'].isin(top_5_clients_ids)]['total_price'].sum()

# Print the results
for client_id in top_5_clients_ids:
    client_spending = client_df[client_df['client_id'] == client_id]['total_price'].sum()
    print(f"Client ID {client_id}: ${client_spending:,.2f}")

```

```python
filtered_client_df = client_df[client_df['client_id'].isin(top_5_clients_ids)]
summary_df = filtered_client_df.groupby('client_id').agg({
    'qty': 'sum',
    'shipping_price': 'sum',
    'total_price': 'sum',
    'line_profit': 'sum'
}).reset_index()

# Convert the sums to millions for readability
summary_df['Total Revenue (Millions)'] = summary_df['total_price'] / 1_000_000
summary_df['Total Shipping Price (Millions)'] = summary_df['shipping_price'] / 1_000_000
summary_df['Total Profit (Millions)'] = summary_df['line_profit'] / 1_000_000

# Rename columns for readability
summary_df.rename(columns={
    'qty': 'Total Units Purchased',
    'shipping_price': 'Total Shipping Price',
    'total_price': 'Total Revenue',
    'line_profit': 'Total Profit'
}, inplace=True)

# Sort by Total Profit (Descending)
summary_df = summary_df.sort_values(by='Total Profit (Millions)', ascending=False)

# Display the summary DataFrame
print(summary_df)
```

```python
# Format the data and rename the columns to names suitable for presentation.
summary_df = pd.DataFrame({
    'client_id': [24741, 33615, 38378, 46820, 66037],
    'Total Units Purchased': [239862, 64313, 73667, 75768, 43018],
    'Total Shipping Price': [7318757.5, 2608560.3, 4895361.6, 2284215.0, 1989501.5],
    'Total Revenue': [8.466399e+07, 9.228995e+06, 1.450805e+07, 1.048972e+07, 1.090884e+07],
    'Total Profit': [3.678278e+07, 2.274105e+06, 3.407426e+06, 2.799759e+06, 3.310009e+06]
})



# Define a function that converts a dollar amount to millions.

def currency_format_millions(value):
    if pd.isna(value):
        return "$0.00"
    return f"${value / 1e6:,.2f}M"


# Define the money columns. 
money_columns = ['Total Shipping Price', 'Total Revenue', 'Total Profit']

# Apply the currency_format_millions function to only the money columns. 
summary_df[money_columns] = summary_df[money_columns].applymap(currency_format_millions)

# Rename the columns to reflect the change in the money format. 

# Rename columns to reflect the change
summary_df.rename(columns={
    'Total Units Purchased': 'Total Units Purchased',
    'Total Shipping Price': 'Total Shipping Price (Millions)',
    'Total Revenue': 'Total Revenue (Millions)',
    'Total Profit': 'Total Profit (Millions)',
}, inplace=True)

# Display the formatted DataFrame
print(summary_df)
```
