```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
client_df['subtotal'] = client_df['unit_price'] * client_df['qty']

client_df.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.

# Define a function to calculate shipping cost based on weight
def calculate_shipping(row):
    if row['unit_weight'] > 50:
        return row['unit_weight'] * 7
    else:
        return row['unit_weight'] * 10

# Create a new column 'shipping_cost' that applies the shipping calculation
client_df['shipping_cost'] = client_df.apply(calculate_shipping, axis=1)

# Display the first few rows to verify the new column
client_df.head()
```

```python
client_df['shipping_price_corrected'] = client_df.apply(
    lambda row: row['unit_weight'] * (7 if row['unit_weight'] > 50 else 10) * row['qty'], axis=1
)
```
